## Visulizing Sales by State and cities

```python
# Region wise sales plot
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(df['Region'],df['Sales'])
plt.title("Region vs Sales")
plt.xlabel("Regions")
plt.ylabel("Sales")
plt.show()
```

```python
# Category wise sales in Each region
plt.figure(figsize=[12,8])
ax=sns.barplot(x="Region",y="Sales",hue='Category',data=df,palette="Blues")
```

```python
# Sub-Category wise sales in Each region
plt.figure(figsize=[12,8])
ax=sns.barplot(x="Region",y="Sales",hue='Sub-Category',data=df,palette="Spectral")
```

```python
plt.figure(figsize=(10,4))
sns.lineplot('Discount','Profit',data=df,color='b',label='Discount')
plt.legend()
```

```python
x=df.iloc[:,[9,10,11,12]].values
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0).fit(x)
    wcss.append(kmeans.inertia_)
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue="Sub-Category",height=5).map(plt.scatter,'Sales','Profit')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='yellow',label='*')
plt.legend()
```
