```python
plt.figure(figsize=(10,5))
sns.barplot(df['Category'],df['Profit'])
```

```python
ds_tech=df[(df['Category']=="Furniture")]
#Let's get the Sales of each Subcategory under Furniture
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Sub-Category", y="Sales", data=ds_tech, palette="magma")
plt.xlabel("categories",fontsize=15)
plt.ylabel("Sales",fontsize=15)
```

```python
ds_tech=df[(df['Category']=="Office Supplies")]
#Let's get the Sales of each Subcategory under Office Supplies
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Sub-Category", y="Sales", data=ds_tech, palette="magma")
plt.xlabel("categories",fontsize=15)
plt.ylabel("Sales",fontsize=15)
```

```python
ds_tech=df[(df['Category']=="Technology")]
#Let's get the Sales of each Subcategory under Technology
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Sub-Category", y="Sales", data=ds_tech, palette="magma")
plt.xlabel("categories",fontsize=15)
plt.ylabel("Sales",fontsize=15)
```

```python
#State wise profit
plt.figure(figsize=[24,15])
ax = sns.barplot(x="State", y="Profit", data=df, palette="Set1",)
plt.xticks(rotation=90, fontsize=16)
plt.yticks(fontsize=15)
plt.title("States VS Profit",fontsize=24)
plt.xlabel("States",fontsize=20)
plt.ylabel("Profit",fontsize=20)
plt.tight_layout()
```

```python
plt.figure(figsize=(15,10))
df['Sub-Category'].value_counts().plot.pie(shadow=True,autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(df['Sub-Category'],df['Profit'])
```

```python
# Region wise sales plot
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(df['Region'],df['Sales'])
plt.title("Region vs Sales")
plt.xlabel("Regions")
plt.ylabel("Sales")
plt.show()
```

```python
# Category wise sales in Each region
plt.figure(figsize=[12,8])
ax=sns.barplot(x="Region",y="Sales",hue='Category',data=df,palette="Blues")
```

```python
# Sub-Category wise sales in Each region
plt.figure(figsize=[12,8])
ax=sns.barplot(x="Region",y="Sales",hue='Sub-Category',data=df,palette="Spectral")
```

```python
#Let's get the Profit of each Subcategory under Technology for each Region
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Sub-Category", y="Profit",hue="Region", data=ds_tech, palette="viridis")
plt.xlabel("Subcategories",fontsize=15)
plt.ylabel("Profit",fontsize=15)
```

```python
grouped= df.groupby("City")
agg_sales=grouped['Sales'].agg(np.sum).sort_values(ascending=False).reset_index()
agg_sales.head()
```

```python
agg_sales.tail()
```

```python
agg_profit=grouped['Profit'].agg(np.sum).sort_values(ascending=False).reset_index()
agg_profit.head()
```

```python
agg_profit.tail()
```

```python
agg_dist=grouped['Discount'].agg(np.sum).sort_values(ascending=False).reset_index()
#Cities with highest aggregated Discount
agg_dist.head()
```

```python
agg_dist.tail()
```
